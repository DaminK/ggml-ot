## Getting started on example data

In this tutorial, we show how to generate a basic synthetic dataset and further work with it.

```python
# import ggml_ot and remaining packages
import ggml_ot

from torch.utils.data import DataLoader
```

#### Generating a dataset

This will create a dataset with three distributions, each with 100 points, where the values of each distribution are shifted vertically to create distinct groups, and plot the element-level embedding:

```python
data = ggml_ot.synthetic_Dataset()
```

For further information on how to adapt the parameters to create different synthetic datasets, check out the `get_pointcloud` function.

Now, to train the model for the global ground metric on the given dataset, call the generated dataset in a Dataloader.

```python
# create training dataset that contains the dataset and meta data like the batching information
training_dataset = DataLoader(data, batch_size=128, shuffle=True)
```

This is how optimal transport performs on the synthetic dataset with the Euclidean distance as a ground metric... 

```python
D_euclidean = data.compute_OT_on_dists(ground_metric="euclidean", plot=True)
```

... and this is how it performs with ggml_ot as the ground metric:

```python
# train the model
# --> learns the global ground metric on the given dataset
# also implicitly plot the embedding and clustermap after 20 iterations
w_theta = ggml_ot.ggml(
    training_dataset, norm=1, max_iterations=20, plot_i_iterations=20, dataset=data
)
```
